# OPERATORS

### Comparison Operators

```python
# Equal (==)
10 == 10
```

```python
10 == 50
```

```python
# Not Equal (!=)
10 != 50
```

```python
"Hanabi" != "Miya"
```

```python
# Less Than (<)
10 < 50
```

```python
# Less Than or Equal to (<=)
10 <= 10
```

```python
# Greater Than (>)
50 > 10
```

```python
# Greater Than or Equal to (>=)
50 >= 10
```

### Comparison Operators

```python
# and
(50 > 10) and (70 > 35)
```

```python
(50 > 10) and (70 > 100)
```

```python
(10 == 10) and ('Hanabi' == 'Hanabi')
```

```python
# or
(100 > 69) or (25 > 75)
```

```python
('Hanabi' > 'Miya') or (10 < 50)
```

```python
# not
not (50 > 40)
```

### Membership Operators

```python
# Example
bini = "Ayokong maniwala, ayokong maniwala eyy!"
```

```python
'eyy' in bini
```

```python
'sheeesh' not in bini
```

```python
'Yo' in bini
```

```python
blackpink = "Blackpink in your area!"
```

```python
blackpink in bini
```

```python
blackpink not in bini
```

```python
'pink' in "Blackpink in your area!"
```
