# FOR LOOPS

#### Performing for loops with List

```python
# Print elements of a list
numbers = [1,2,3,4,5]

for number in numbers:
    print(number)
```

```python
# Used as a counter. Print any string, not necessary the elements of the list
for number in numbers:
    print('Eyyy')
```

```python
# Performing operations within the loop
for Meow in numbers:
    print(Meow + Meow)
```

#### Performing for loops with dictionaries

```python
# Example
ice_cream_dict = {'Name':'Kath', 'Weekly intake':5, 'favorite_ice_cream':['Matcha', 'Chocolate']}
type(ice_cream_dict)
```

```python
# Print all keys in the dictionary
for cream in ice_cream_dict.keys():
    print(cream)
```

```python
# By default, this will print all keys
for cream in ice_cream_dict:
    print(cream)
```

```python
# To print the values
for cream in ice_cream_dict.values():
    print(cream)
```

```python
# To print keys and values
for key, value in ice_cream_dict.items():
    print(key, '->', value)
```

```python
# Print items, setting 1 variable. Python will treat that variable as tuple and output the key-value pair
for item in ice_cream_dict.items():
    print(item)
```

#### Nested For Loops

```python
# Example
flavors = ['Vanilla', 'Chocolate', 'Cookies and Cream']
toppings = ['Hot Fudge', 'Oreos', 'Marshmallows']

for one in flavors:
    for two in toppings:
        print(one, 'topped with', two)
```
