# IF - ELIF - ELSE STATEMENTS

#### IF STATEMENT

```python
# If True, will run the block
if 25 > 10:
    print('It worked!')
```

```python
# If False, no output and skips the block
if 25 < 10:
    print('It worked!')
```

#### IF - ELSE STATEMENT

```python
# If True, will run the block and skip else 
if 25 > 10:
    print('It worked')
else: 
    print('It did not work...')
```

```python
# If False, will skip the if block and run the else block
if 25 < 10: 
    print('It worked')
else:
    print('It did not work...')
```

#### ELIF STATEMENT

```python
# We can have multiple elif statements in one block
if 25 < 10:
    print('It worked!')
elif 25 < 20:
    print('Elif worked!')
elif 25 < 21:
    print('Elif2 worked!')
elif 25 < 40:
    print('Elif3 worked!')
elif 25 < 50:
    print('Elif4 worked!')
else: 
    print('It did not work...')
```

```python
# We can have if and elif and no else statement in the block
if 25 < 10:
    print('It worked!')
elif 25 < 20:
    print('Elif worked!')
elif 25 < 21:
    print('Elif2 worked!')
elif 25 < 40:
    print('Elif3 worked!')
elif 25 < 50:
    print('Elif4 worked!')
```

##### LOGICAL OPERATORS ON IF-ELIF-ELSE STATEMENTS

```python
# or 
if (25 < 10) or (25 < 50): 
    print('It worked!')
elif 25 < 20:
    print('Elif worked!')
else:
    print('It did not work...')
```

```python
# and
if (25 < 10) and (25 < 50): 
    print('It worked!')
elif 25 < 20:
    print('Elif worked!')
else:
    print('It did not work...')
```

##### IF-ELSE STATEMENT IN ONE LINE

##### NESTED IF STATEMENTS

```python
# Example
if 25 > 10:
    print('It worked!')
    if 5 > 1:
        print('This nested if statement worked!')
elif 25 < 20:
    print('Elif worked!')
elif 25 < 21:
    print('Elif2 worked!')
elif 25 < 40:
    print('Elif3 worked!')
elif 25 < 50:
    print('Elif4 worked!')
```
