```python
# Assign multiple values to multiple variables at once
x,y,z = 'Aiah','Mikha','Sheena'
print(x)
print(y)
print(z)
print(x,y)
print('My biases in bini:',x ,',', y, ',', z)
```

```python
# Assign one value to multiple variables at once
x = y = z = 'Mabuhay and we are bini'
print(x)
print(y)
print(z)
```

```python
# Assing list to a variable
other_bini_members = ['Colet','Jhoanna', 'Stacey', 'Gwen', 'Maloi']
a, b, c, d, e = other_bini_members
print(a)
print(b)
print(c)
print(d)
print(e)
```

```python
# Naming variables

# Camel Case
testVariableCase = 'eyyKaMuna'
print(testVariableCase)

# Pascal Case
TestVariableCase = 'EyyKaMuna'
print(TestVariableCase)

# Snake Case
test_variable_case = 'eyy_ka_muna'
print(test_variable_case)
```

```python
# Operators in variable. + for string concatenation
x = 'Ice Cream is my favorite' + ' love joy hope'
print(x)
```

```python
# We cannot concateate str and number
x = 'Hello' + 69
print(x)
```

```python
# We can use operator if same number data type
x = 10 + 20
print(x)
```

```python
# Concatenate multiple strings and multiple numbers
a = 'hi'
b = ' hey'
c = ' hello'
print(a+b+c)

e = 5
f = 2
g = 3
print(e+f+g)
```
