# WHILE LOOPS

```python
# Example
number = 0
while number < 5:
    print(number)
    number = number + 1
```

#### Break statement in a while loop

```python
# Example
number = 0
while number < 5:
    print(number)
    if number == 3:
        break
    number = number + 1
```

#### Else statement in a While loop

```python
# Example
number = 0
while number < 5:
    print(number)
    if number == 6:
        break
    number = number + 1
else:
    print('No longer < 5')
```

```python
# Else statement will not execute if the while loop was stopped due to the break statement
number = 0
while number < 5:
    print(number)
    if number == 3:
        break
    number = number + 1
else:
    print('No longer < 5')
```

#### Continue Statement in a While loop

```python
# Example
number = 0
while number < 5:
    number = number + 1
    if number == 3:
        continue
    print(number)
else:
    print('No longer < 5')
```
